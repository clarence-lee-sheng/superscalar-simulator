```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd 
```

```python
df = pd.DataFrame(
    {
    'Benchmark': ["Vector addition", "Matrix Multiplication", "Odd counts", "Decision Tree", "Convolution"], 
    'Static Not Taken': [61.5,50.0,49.7, 57.1, 49.6],
    'Static Taken': [95.1,56.9,81.7, 82.3,63.8], 
    'One Bit': [95.1,88.9,65.7, 89.7,87.4], 
    'Two Bit Saturating Counter': [95.1,94.2,81.7, 90.5, 92.1],
    'Perceptron': [95.1,94.2,90.6, 91.5, 92.1]
    }
)
```
